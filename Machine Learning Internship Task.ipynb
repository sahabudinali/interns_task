```python
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
len(X)
```

```python
len(y)
```

```python

seed=7
test_size=0.33
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=test_size,random_state=seed)
```

```python
model=XGBClassifier()
model.fit(X_train,y_train)
```

```python
print(model)
```

```python
y_pred=model.predict(x_test)
```

```python
prediction_Score=[round(value) for value in y_pred]
```

```python
accuracy_score=accuracy_score(y_test,prediction_Score)
```

```python
print("Accuracy Score of XBBOOST model is : ", accuracy_score*100)
```
